#Diabetes Dataset


##Summary of Dataset
This data set is for predicting if the patient has diabetes based on certain measurements included in the dataset.
It includes only females of at least 21 years old of Pima heritage.

There are  9 columns and 768 data entries

Pregenacies - Number of pregnancies

Glucose - Glucose level in blood

BloodPressure - Measured bloodpressure

SkinThickness - Thickness of skin

Insulin - Insulin level in blood

BMI - Body mass index

DiabetesPedigree Function - To express Diabetes Percentage

Age - Age of person

Outcome - From the measured diagnosis give final result as 1 if Yes diabetes and 0 if No

link for dataset - https://www.kaggle.com/datasets/akshaydattatraykhare/diabetes-dataset?select=diabetes.csv

##Objectives for the Dataset
1. Get the Info of the dataset
2. Count the People who have diabetes and those who donot.
3. Group according to the age and BMI.
4. Filter according to the no. of pregnancies.
5. Check for empty data entries.
6. Calculate mean age for people with diabetes.

## Counting and Grouping data

```python
# Checking the number of people with diabetes and without diabetes

outcome = (df['Outcome'] == 1).value_counts()
print(outcome)
outcome.plot.pie()
```

```python
# Grouping according to age and BMI in ascending order

df[['Age', 'BMI']].sort_values(by='Age', ascending=True).groupby(['Age', 'BMI']).size()
# result = df.groupby(['Age', 'BMI']).size().sort_index(level='Age')
# result
```

```python
# According to pregnancies

(df['Pregnancies'] > 4).value_counts()
```

```python
# According to age

age = df[df['Age'] < 50]
age.sort_values(by = 'Age')
```

##Filtered Dataset for People with diabetes

```python
filt_diabetes = (df['Outcome'] == 1)
df.loc[filt_diabetes, 'Age'].mean()
```

```python
df_withDiabetes = df[filt_diabetes]
mean = df_withDiabetes.mean()
print(mean)
mean.plot.bar()
```

```python
median = df_withDiabetes.median()
print(median)
median.plot.bar(color='green')
```

```python
filt_age = df['Age'] < 40
df_withDiabetes.loc[filt_age,'Pregnancies'].value_counts()
```
